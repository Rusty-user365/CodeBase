#Sentiment Analysis

```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
result=classifier("I'm bored i lost the love of my life, I'm jobless , I feel alone")

print(result)
```

```python
pipeline(task="sentiment-analysis")("I Am very happy   ")
```

```python
pipeline(task="sentiment-analysis")\
 ("No more tech hiring in India, \
 Donald Trump tells Google,\
  Microsoft and others to focus on Americans")
```

```python
pipeline(task="sentiment-analysis",model="facebook/bart-large-mnli")\
 ("No more tech hiring in India, \
 Donald Trump tells Google,\
  Microsoft and others to focus on Americans")
```

```python
from transformers import pipeline
pipeline(task="sentiment-analysis", model="tabularisai/multilingual-sentiment-analysis")(
    "No more tech hiring in India, Donald Trump kills Google, Microsoft and others to kill Americans"
)
```

```python
from transformers import pipeline
classifier=pipeline(task="sentiment-analysis", model="SamLowe/roberta-base-go_emotions")


task_list=["i really hate  ethics ",\
    " how will Donald Trump kills Google?", \
           " Microsoft and others to progress and make  Americans great again",\
    " Love my family."]
classifier(task_list)
```

#Text Generation

```python
#use a pipeline as a high-level helper
from transformers import pipeline
text_generator=pipeline("text-generation",model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B")
generated_text=text_generator("Today is a rainy day in London",
                              truncation=True,
                              num_return_sequences=2)
print("Generated_text:\n",generated_text[0]['generated_text'])
```
