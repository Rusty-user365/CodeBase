```python
import numpy as np
import tensorflow  as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['age','affordability']],df.bought_insurance,test_size=0.2,random_state=25)
```

```python
len(X_train)
```

```python
X_train_scaled=X_train.copy()#made copy of dataFrame
X_train_scaled['age']=X_train_scaled['age']/100

X_test_scaled=X_test.copy()#made copy of dataFrame
X_test_scaled['age']=X_test_scaled['age']/100


```

```python
X_train_scaled
```

```python
import time
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
start_time1 = time.time()

model.fit(X_train_scaled, y_train, epochs=5000)
end_time1 = time.time()
```

```python
# Get weights and bias from the single Dense layer
weights, bias = model.layers[0].get_weights()

print("Final weights:", weights.flatten())
print("Final bias:", bias)
```

```python
model.evaluate(X_test_scaled,y_test)
```

```python
#train data is small and simple  can overfit with multi layer model but trains fast and accurate
import time
model2 = keras.Sequential([
    keras.Input(shape=(2,)),  # Input layer
    keras.layers.Dense(4, activation='relu'),  # Hidden layer
    keras.layers.Dense(1, activation='sigmoid')  # Output layer
])

model2.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
start_time2 = time.time()

model2.fit(X_train_scaled, y_train, epochs=500, verbose=1)
end_time2 = time.time()
```

```python
# Layer 0: Dense(4, activation='relu')
weights_hidden, bias_hidden = model2.layers[0].get_weights()

# Layer 1: Dense(1, activation='sigmoid')
weights_output, bias_output = model2.layers[1].get_weights()
```

```python
print("Hidden layer weights:\n", weights_hidden)
print("Hidden layer bias:\n", bias_hidden)
print("Output layer weights:\n", weights_output)
print("Output layer bias:\n", bias_output)
```

```python
model2.evaluate(X_test_scaled,y_test)
```

```python
#using function
import math
epsilon=1e-15
def logloss(y_predicted,y_true):
    total_error=0
    for yp,yt in zip(y_predicted,y_true):
     # Clip prediction to avoid log(0)
        yp=max(epsilon,min(yp,1-epsilon))
        total_error+=yt*math.log(yp)+(1-yt)*math.log(1-yp)
    logloss=-total_error/len(y_true)
    return logloss
```

```python
import numpy as np

def sigmoid_numpy(x):
    return 1 / (1 + np.exp(-x))  

# Test with a NumPy array
sigmoid_numpy(np.array([12, 0, 1]))
```

```python
#this code is for neural network in ml it was for linear regression or simple logistic regression
def gradient_descent(age,affordability,y_true,epochs,loss_thresold):
    #w1,w2,bias
    w1=0
    w2=0
    bias=-3
    rate=0.5
    n=len(age)
    for i in range(epochs):
        weighted_sum=w1*age+w2*affordability+bias#this is just  linear regression
        y_predicted=sigmoid_numpy(weighted_sum)# this is just sigmoid
        loss=logloss(y_predicted,y_true)
        #Gradient calculation
        w1d=(1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
        w2d=(1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))#Gradient computed with respect to corresponding feature
        bias_d=np.mean(y_predicted-y_true)

        w1=w1-rate*w1d
        w2=w2-rate*w2d
        print(f'Epoch: {i},w1:{w1},w2:{w2}, bias:{bias}, loss{loss}')
        if loss<=loss_thresold:#to stop when loss is same as tensorflow function to compare weights
            break
    return w1,w2,bias         
```
